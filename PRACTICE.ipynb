```python
 import pandas as pd
import numpy as np
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
import matplotlib.pyplot as plt
```

```python
car=pd.read_csv("/content/Cars.csv")
car.head()
```

```python
car.isna().sum()
```

```python
car.describe()
```

```python
sns.distplot(x=car['WT'])
plt.title("Distribution of weight of vehicles in thousand pounds")
plt.xlabel('Weight of vehicles in thousand pounds')
plt.show()
```

```python
sns.boxplot(x=car['MPG'])
plt.title("Distribution of MPG of vehicles")
plt.show()
```

```python
sns.boxplot(x=car['HP'])
```

```python
sns.scatterplot(x=car['MPG'],y=car['HP'])
plt.title("MPG vs HP")
plt.show()
```

```python
sns.barplot(x=car['HP'],y=car['WT'])
plt.title("HP vs WT")
plt.show()
```

```python
sns.scatterplot(x=car['HP'],y=car['WT'])
plt.title("HP vs WT")
plt.show()
```

```python
sns.scatterplot(x=car['HP'],y=car['VOL'])
plt.title("HP vs VOL")
plt.show()
```

```python
sns.scatterplot(x=car['HP'],y=car['SP'])
plt.title("HP vs SP")
plt.show()
```

```python
sns.scatterplot(x=car['WT'],y=car['SP'])
plt.title("WT vs SP|")
plt.show()
```

```python
sns.scatterplot(x=car['WT'],y=car['VOL'])
plt.title("WT vs VOL")
plt.show()
```
